List is an ordered,mutable,collectioms of items.It allows duplicate values and can store different types of data types.
supports slicing.

ex: fruits=['apple','banana','mango']

Functions:
append() add items at end of the list
insert() insert items at specific position
remove() remove item using index num
sort() sort list
reverse() reverse the list
len() count elements
count() counts number of occurances

```python
stud=['meghana','teju','sneha']
print(stud)
print(type(stud))
stud.append('deepu')
print(stud)
stud.insert(2,'chikki')
print(stud)
stud.remove('sneha')
print(stud)
stud.sort()
print(stud)
stud.reverse()
print(stud)
l=len(stud)
print(l)
```

Functions:

set:
a set id unordered collection of unique elements.
mutable and no dulipcates are allowed in sets
ex: num={1,2,3,4,5}

funvtions:
add() add element
remove() removes element
union() combine sets
difference() unique elements
clear() removes all the elements from the set

```python
n1={'chikki','chinni','teju'}
n2={'meghna','anu','chikki'}
n1.add('sneha')
print(n1)
n1.remove('chinni')
print(n1)
unions=n1.union(n2)
print(unions)
dif=n1.difference(n2)
print(dif)
inter=n1.intersection(n2)
print(inter)
n2.clear()
print(n2)
```

dict:
stores data in key value pairs.
mutable keys must be unique
stores related information together


functions:
keys() returns all keys
values() returns all values
items() returns list of tuple of keyvalue pairs
get() access the values
pop() remove keys and prints the removed value
update() update the values in dict

```python
stud={
    'name':'chikki',
    'age':22,
    'course':'python'
}
print(stud.keys())
print(stud.values())
print(stud.items())
print(stud.get('course'))
stud.pop('course')
print(stud)
stud.update({'course':"java"})
print(stud)
stud.update({'age':23})
```
